# **1. Handling Missing Values**

```python
features_nan=[feature for feature in data.columns if data[feature].isnull().sum()>1]
for feature in features_nan:
  print("{}: {}% missing values".format(feature,np.round(data[feature].isnull().mean())))
features_nan
```

```python
def replace_feature(data,features_nan):
  dataset=data.copy()
  dataset[features_nan]=dataset[features_nan].fillna("missing")
  return dataset
data=replace_feature(data,features_nan)
data[features_nan].isnull().sum()
```

# 2. Handling the Outliers

```python
maximum_threshold=data['PassengerId'].quantile(0.95)
maximum_threshold
```

```python
data[data['PassengerId']> maximum_threshold]
```

```python
minimum_threshold=data['PassengerId'].quantile(0.05)
minimum_threshold
```

```python
data[data['PassengerId']< minimum_threshold]
```

```python
data[(data['PassengerId']< maximum_threshold) & (data['PassengerId']> minimum_threshold)]
```
